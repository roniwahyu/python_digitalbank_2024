```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'review_bank_digital_googleplaystore_cleaned_text_preprocessing.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
# print(data.head())
data
# Assuming the columns are named 'content' and 'cleaned_content_multi'
# Check for NaN values in 'content' and 'cleaned_content_multi' columns
print(data[['content', 'cleaned_content_multi']].isnull().sum())
```

```python
# Filter the data to include only rows where 'cleaned_content_multi' is not null
data = data.dropna(subset=['cleaned_content_multi'])
```

```python

# Calculate the length of the content for both original and cleaned content
data['original_length'] = data['content'].apply(len)
data['cleaned_length'] = data['cleaned_content_multi'].apply(len)
```

```python
data
```

```python
filename = 'hasil/001_2_review_bank_digital_googleplaystore_cleaned_counted.csv'
data.to_csv(filename, index=False)
```

```python


# Plot histograms to compare the distributions
plt.figure(figsize=(12, 6))

# Histogram for original content length
plt.subplot(1, 2, 1)
plt.hist(data['original_length'], bins=50, color='blue', alpha=0.7, label='Original Content')
plt.xlabel('Length of Content')
plt.ylabel('Frequency')
plt.title('Distribution of Original Content Length')
plt.legend()

# Histogram for cleaned content length
plt.subplot(1, 2, 2)
plt.hist(data['cleaned_length'], bins=50, color='green', alpha=0.7, label='Cleaned Content')
plt.xlabel('Length of Content')
plt.ylabel('Frequency')
plt.title('Distribution of Cleaned Content Length')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
